7. Find how many null values, missing values are present. Deal
with them. Print out how many rows have been modified.

```python
df_movies=pd.read_csv('movies.csv')
print("Null Values  handled in movies {}".format(df_movies.isnull().count().sum()))
print("Null Values handled in tags {}".format(df.isnull().count().sum()))
print("Null Values handled  in ratings {}".format(df_ratings.isnull().count().sum()))
df_movies.drop(df_movies.isnull().index,inplace=True)
df.drop(df.isnull().index,inplace=True)
df_ratings.drop(df_ratings.isnull().index,inplace=True)
print("Null Values  in movies {}".format(df_movies.isnull().count().sum()))
print("Null Values  in tags {}".format(df.isnull().count().sum()))
print("Null Values   in ratings {}".format(df_ratings.isnull().count().sum()))
```

10. Perform an inner join of movies and tags based on movieId.

```python
merge_val=df_movies.merge(df,on='movieId',how='inner').drop_duplicates('movieId')
#merge_val=merge_val.drop_duplicates('movieId')
merge_val
```

 11. Print out the 5 movies that belong to the Comedy genre and
have rating greater than 4.

```python
df_rat_dup=df_ratings.drop_duplicates('movieId')
rating_result=merge_val.merge(df_rat_dup[['movieId','rating']],on='movieId',how='inner')
rating_result_final=rating_result[rating_result.rating > 4]
print("5 movies that belong to the Comedy genre and have rating greater than 4 : ")
for i in rating_result_final['title'].head(5):
    print(i)
```
